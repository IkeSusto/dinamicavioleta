Eu quero montar uma dinâmica para fazer com alunos em sala de aula para mostrar os limites e possibilidades do PIB. 

A ideia é criar um jogo em flash (vou subir via github por um repositório). Duas telas, uma compartilhada por todos que vai ser projetada, e o celular (cada jogador acessará via celular o seu próprio ambiente durante o jogo)
Em um primeiro momento use placeholders para os textos, objetos, fncionalidades e afins. Vamos programando uma coisa de cada vez e muito bem encapsulada para evitar problemas. Sempre que vc achar que mais detalhes ajudaria a melhorar a tua respota me avise que explico melhor. 
Vai ter uma página inicial com um Qr code para os alunos se conectarem ao jogo via celualar. 
Ao conectar, cada jogador vai escolher um nick e selecionar qual país deseja representar, dentre 4 países produtores de biscooitos (cada um com uma cor). A ideia é copiar o esquema do quizziz, menti e outros sites que fazem isso.
Após todos se conectarem o jogo começa. Vai ter um timer de 20s com um texto informativo explicando a primeira etapa. A primeira etapa consiste em uma versão do jogo dalgona (squidgame). Cada jogador vai receber uma das formas de forma meio aleatória e tem que traçar o desenho dela corretamente na tela para cortar o biscoito. Se cortar bem (acima de 95% de acurácia) o biscoito passa pelo setor de qualidade e é comercializado, senão ele quebra e vc perde a produção. Os alunos vão ficar nisso por 3 minutos, tentando produzir o máximo de biscoitos possíveis. Os biscoitos terão corolação igual a do país mas ainda com cor de comida.
Para definir o próximo biscoito de cada jogador, o programa vai rolar um dado e cada país vai ter chances diferentes para cada tipo de biscoito (Aqui quero representar desigualdade).
Acabada essa primeira fase o flash vai mostrar uma página de sumarização do primeiro ano de prudução e vai calcular o valor do PIB de cada país usando a fórmula: GDP (Y) is the sum of consumption (C), investment (I), government expenditures (G). Y = C + I + G. 
Isso deverá ser mostrado na tela principal. 
Nesse primeiro momento vc vai rodar mais dados. O custo do governo vai ser na sorte com mensagens irônicas, no estilo de manchetes de jornal. Depois vc me sugere algumas. A ideia aqui é representar o quão mais e menos cada país tem que gastar com infraestrutura, dado seu histórico de dominação, acúmulo de capital, conhecimento, tecnologia, estabilidade social e financiera, ambiental, civil e guerra e afins. A segunda etapa, logo que terminar a primeira vai ser o jornal do ano (tela principal), com essas mensagens. Essas mensagens vão trazer os custos governamentais de cada país assim como oportunidades de investimento externo (fatores que vão influenciar no calculo do PIB de cada país). Depois vamos mostrar na tela pricipal a tabela com o PIB de cada país.
A tarceira etapa vai ser uma espécie de leilão de qualidade de vida onde os grupos poderão gastar o próprio custo governamental em features para facilitar a produção própria ou atrapalhar a dos colegas (pq o capitalismo é cruel). Aqui temos varias ideias de autocliquer e poderzinhos de minigames para nos inspirarmos. Quero uma feature que corte o biscoito sozinho, so que de forma mais lenta que uma pessoa conseguiria, uma que estabilize mais o touch pro formato do desenho para ela acertar com mais facilidade os cortes e uma que venda biscoito quebrado como subproduto para agregar ao PIB mesmo as falhas, outras agt pensa junto. 
Vamos repetir essa sequência de 3 etaoas 3 vezes (3 rodadas/ 3 anos).
Quero manipular as condições para que um dos países faça muito pouco trabalho e consiga os poderzinhos com facilidade (simulando países que conseguem ter alta qualidade de vida sem sacrificar o PIB) e um outr país que só se ferra e tem que ficar penando em todas as rodadas pra acompanhar o desenvolvimento dos outros.  

Meu primeiro pedido nesse grande projeto é: Me ajude a reescrever esse roadmap de forma completa e quebrada de forma inteligente para podermos codarmos juntos. 

Aqui está o roadmap reescrito e organizado em etapas claras e encapsuladas para facilitar o desenvolvimento:

---

## **Roadmap do Projeto**

### **1. Configuração inicial**
- **Objetivo:** Criar a estrutura básica do projeto e preparar o ambiente de desenvolvimento.
- **Tarefas:**
  1. Configurar o repositório no GitHub.
  2. Escolher as tecnologias para o desenvolvimento:
     - **Frontend:** HTML5, CSS, JavaScript (substituindo Flash, que está obsoleto).
     - **Backend:** Node.js ou Python (para gerenciar conexões e lógica do jogo).
     - **Comunicação:** WebSocket (para comunicação em tempo real entre os celulares e a tela principal).
  3. Criar placeholders para textos, objetos e funcionalidades.

---

### **2. Tela inicial**
- **Objetivo:** Criar a página inicial com o QR Code para os alunos se conectarem ao jogo.
- **Tarefas:**
  1. Criar uma página inicial com um QR Code gerado dinamicamente.
  2. Configurar o backend para gerar links únicos para cada sessão do jogo.
  3. Implementar a lógica para os jogadores se conectarem via celular:
     - Cada jogador insere um nick.
     - Cada jogador escolhe um país (entre 4 opções).

---

### **3. Primeira etapa: Produção de biscoitos**
- **Objetivo:** Implementar o minigame de produção de biscoitos.
- **Tarefas:**
  1. Criar o timer de 20 segundos com texto explicativo na tela principal.
  2. Implementar o jogo de corte de biscoitos:
     - Cada jogador recebe uma forma aleatória.
     - Implementar a lógica de desenho e cálculo de acurácia (acima de 95% para sucesso).
     - Adicionar lógica para biscoitos quebrados (falha na produção).
  3. Configurar o sistema de cores:
     - Cada país tem uma cor específica para os biscoitos.
  4. Implementar a lógica de desigualdade:
     - Rolar um dado para definir o próximo biscoito.
     - Cada país tem chances diferentes para cada tipo de biscoito.

---

### **4. Sumarização do primeiro ano**
- **Objetivo:** Calcular e exibir o PIB de cada país.
- **Tarefas:**
  1. Implementar a fórmula do PIB: `Y = C + I + G`.
  2. Exibir os resultados na tela principal:
     - Mostrar o número de biscoitos produzidos e comercializados.
     - Mostrar o valor do PIB de cada país.
  3. Rodar dados para custos governamentais e oportunidades de investimento externo:
     - Gerar mensagens irônicas no estilo de manchetes de jornal.
     - Exibir os custos e oportunidades na tela principal.

---

### **5. Segunda etapa: Jornal do ano**
- **Objetivo:** Mostrar os custos governamentais e oportunidades de investimento externo.
- **Tarefas:**
  1. Criar uma tela de jornal com mensagens irônicas.
  2. Exibir os custos governamentais e oportunidades de investimento externo para cada país.
  3. Atualizar os valores do PIB com base nos custos e investimentos.

---

### **6. Terceira etapa: Leilão de qualidade de vida**
- **Objetivo:** Permitir que os jogadores gastem recursos para melhorar sua produção ou atrapalhar os colegas.
- **Tarefas:**
  1. Implementar o sistema de leilão:
     - Cada jogador pode gastar recursos para comprar melhorias ou sabotagens.
  2. Criar as seguintes funcionalidades:
     - **Autoclicker:** Produz biscoitos automaticamente, mas de forma mais lenta.
     - **Estabilizador de touch:** Facilita o corte dos biscoitos.
     - **Venda de biscoitos quebrados:** Agrega ao PIB mesmo as falhas.
  3. Exibir os resultados do leilão na tela principal.

---

### **7. Repetição das rodadas**
- **Objetivo:** Repetir as etapas anteriores por 3 rodadas (3 anos).
- **Tarefas:**
  1. Manipular as condições para simular desigualdade:
     - Um país tem alta qualidade de vida com pouco esforço.
     - Outro país enfrenta dificuldades em todas as rodadas.
  2. Atualizar os valores do PIB e exibir os resultados após cada rodada.

---

### **8. Finalização**
- **Objetivo:** Mostrar os resultados finais e encerrar o jogo.
- **Tarefas:**
  1. Exibir a tabela final com o PIB acumulado de cada país.
  2. Mostrar mensagens de encerramento e reflexões sobre desigualdade e qualidade de vida.

---
